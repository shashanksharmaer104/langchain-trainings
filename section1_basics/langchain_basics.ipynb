# Interacting with local LLM

```python
from langchain_ollama import ChatOllama
import os

llm = ChatOllama(
    base_url="http://localhost:11434",
    model="gemma3:1b", #llama3.2:latest
    temperature=0.5,
    max_tokens=250,
)

response = llm.invoke("Hello, how are you doing today?")

print(response)
```

# Interacting with local cloud LLM

```python
from langchain_ollama import ChatOllama
import os

llm = ChatOllama(
    base_url="http://localhost:11434/",  # Ollama cloud endpoint
    model="qwen3.5:cloud",
    temperature=0.5,
    max_tokens=250,
    headers={
        "Authorization": f"Bearer {os.getenv('OLLAMA_CLOUD_API_KEY')}"  # Cloud auth
    }
)

response = llm.invoke("Hello, how are you doing today?")

print(response)

print(response.content)
```

## Prompt & Chat Template

#### Prompt Template

```python
from langchain_core.prompts import PromptTemplate


prompt_template = PromptTemplate.from_template("What is the advantage of running the LLM in {env}?")

prompt = prompt_template.invoke({"env": "local machine"})

print(prompt)

content = llm.invoke(prompt).content

print(content)
```

#### Chat Prompt Template

```python
from langchain_core.prompts import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,)

systemMessage = SystemMessagePromptTemplate.from_template("You are an LLM expert")
humanMessage = HumanMessagePromptTemplate.from_template("What is the advantage of running the LLM in {env}?")

# 1st way
prompt_template = ChatPromptTemplate([
    systemMessage,
    humanMessage
])

# 2nd way
# prompt_template = ChatPromptTemplate([
#     ("system", "You are an LLM expert"),
#     ("human", "What is the advantage of running the LLM in {env}?")
# ])

prompt_template

prompt = prompt_template.invoke({"env": "local machine"})

# print(prompt)

content = llm.invoke(prompt).content

print(content)
```

#### MessagePlaceHolder

```python
from langchain_core.prompts import (
    ChatPromptTemplate,
    MessagesPlaceholder)

from langchain_core.messages import HumanMessage

prompt_template = ChatPromptTemplate([
    ("system", "You are an LLM expert"),
    MessagesPlaceholder("msg"),
])

prompt = prompt_template.invoke({"msg": [HumanMessage("What is the advantage of running the LLM in local machine?")]})

content = llm.invoke(prompt).content

print(content)
```
